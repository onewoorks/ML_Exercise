```python
import pandas as pd
from sklearn import linear_model

df = pd.read_csv("hiring.csv")
df
```

```python
# fill empty exprience with zero
df.experience = df.experience.fillna('zero')
df
```

```python
# convert word into number
from word2number import w2n
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
# fill empy test score with media score
import math
median_score = math.floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']], df['salary($)'])
```

```python
# predict salary for 2 years experience, 9 test score and 6 interview score
predict_salary = reg.predict([[2,9,6]])
print("Predicted Salary : {}".format(predict_salary[0]))
```

```python
# predict salary for 12 years experience, 10 test score, 10 interview score
predict_salary = reg.predict([[12,10,10]])
predict_salary
```
